## データの読み込み
- train.csvを読み込みましょう
- 読み込んだものは変数trainに代入しましょう

```python
train = pd.read_csv('train.csv')
```

## trainの先頭行を見てみましょう

```python
train.head()
```

## 基本統計量を確認してみましょう
- describe関数を使います

```python
# describe関数は数値のデータのみ、様々な計算結果を表示する。
## stdは標準偏差、50%は中央値を表している
train.describe()
```

## データの型を確認してみましょう
- info関数を使います

```python
train.info()
```

## 1つのカラムだけに注目してみましょう
- ここでは"y"というカラムを見てみましょう

```python
train['y'].headad()
```

## yの平均と中央値を見てみましょう
- 平均値はmean関数、中央値はmedian関数を使います

```python
train['y'].mean()
```

```python
train['y'].median()
```

## yの値が150以上のデータのみを見てみましょう

```python
train[train['y'] >= 150].head()
```

## 曜日が月となっているデータのみを見てみましょう

```python
train[train['week'] == '月'].head()
```

## 曜日が火となっているデータをyで昇順・降順にして見てみましょう
- ソートはsort_values関数を使います。sort_values(by="XXX")と書きます
- 降順にしたい場合はカッコの中にオプションとしてascending=Falseを書きます

```python
# 昇順
train[train['week'] == '火'].sort_values(by="y").head()
```

```python
# 降順
train[train['week'] == '火'].sort_values(by="y", ascending=False).head()
```

## 曜日が月曜日の時のyの平均値を見てみましょう
- 今までの応用です。まず、trainから曜日が月曜日の条件で抜きだし、yを選択した上で、その後に平均値を求める関数を使います

```python
train[train['week'] == '月']['y'].mean()
```
