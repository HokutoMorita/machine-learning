```python
# y = 1 / (1 + exp(-x))
import numpy as np
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
```

```python
# シグモイド関数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Relu関数
def relu(x):
    return np.maximum(0, x)
```

```python
# -6から6までの値を0.1刻みで格納する。
x = np.arange(-6.0, 6.0, 0.1)
```

```python
y = sigmoid(x)
```

```python
plt.plot(x, y)
```

```python
y_relu = relu(x)
```

```python
y_relu
```

```python
plt.plot(x, y_relu)
```

```python
x = np.array([0.6, 0.4])
w = np.array([0.2, 0.8])
```

```python
theta = -0.2
```

```python
# np.dot()は行列の内積（要素の掛け算の総和）
input = np.dot(x, w) + theta
```

```python
input
```

```python
output = sigmoid(input)
```

```python
output # outputの値は1を超えてたらおかしい（シグモイド関数は0から1であるため）
```
